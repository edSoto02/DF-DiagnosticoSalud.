### Regresión Logístoca

```python
X = dataframe.drop('diagnosis', axis=1)
y= dataframe['diagnosis']

X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.3, random_state=42)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
print("Precisión del conjunto de entrenamiento: ", model.score(X_train, y_train))
print("Precisión del conjunto de pruebas: ", model.score(X_test, y_test))

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
matriz_confusion = confusion_matrix(y_test, y_pred)
print('Presición del modelo: {:.2f}%'.format(accuracy*100))

print('Reporte de Clasificación: ')
print(classification_report(y_test, y_pred))

print('Matriz de Confusión: ')
print(confusion_matrix(y_test, y_pred))

plt.figure(figsize= (5,5))
sns.heatmap(matriz_confusion, annot=True, fmt='g', cmap='Blues')
```

### Interpretcación de los resultados para Regresión Logística.

Presisción del modelo.
 * La presición del del conjunto de entrenamineto es del 87.92%
 * La presición dle conjunto de pruebas es del 75.56%

Reporte de clasificación.
 * Presición: La presición para la clase 0 es del 75%, lo que significca que el 75% de las predicciones clasificadas como clase 0 fueron correctas. Para la clase 1, la presición el del 76%.
 * Recall: El Recall para la clase 0 es del 82%, lo que significa que el 82% de las instancias de la clase 0 fueron correctamente identificadas por el modelo. Para la clase 1, el Recall es del 68%.
 * F1-score: El f-score para la clase  es del 78%, que es una medida de la presión y el recall ponderados. Para la clase 1, el f1-score es del 72%. 
 * Support: El soporte indica el número de ocurrencias de cada clase en el conjunto de pruebas.

Matriz de confución.
 * Muestra que el modelo predijo correctamente 40 instancias de la clase 0 y 28 de la clase 1.
 * Predijo incorrectamente 9 instancias de la clase 0 como clase 1 y 13 instancias de la clase 1 como clase 0.

En resuemen, el modelo de regresión logística tiene una precisión razonable en el conjunto de pruebas del 75.56%.
Hay algunas instancias de falsos positivos y falsos negativo, como se muestra en la matriz de confusión.
Estos resultados indican que el modelo puede tener marjen de mejora, pero en general es capaz de clasificar las instancias con una precisión aceptable. 

### Clasificador de árbol de decisión

```python
X = dataframe.drop('diagnosis', axis=1)
y= dataframe['diagnosis']

X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.3, random_state=42)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
print("Precisión del conjunto de entrenamiento: ", model.score(X_train, y_train))
print("Precisión del conjunto de pruebas: ", model.score(X_test, y_test))

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
matriz_confusion = confusion_matrix(y_test, y_pred)

print('Presición del modelo: {:.2f}%'.format(accuracy*100))

print('Reporte de Clasificación: ')
print(classification_report(y_test, y_pred))

print('Matriz de Confusión: ')
print(confusion_matrix(y_test, y_pred))
sns.heatmap(matriz_confusion, annot=True, fmt='g', cmap='Greens')
```

### Interpretcación de los resultados para Arbol de Decisión.

Prescisión del modelo.
 * La precisión del conjunto de entrenamiento es del 100%
 * La precisión del conjunto de pruebas es del 67.78%.

Reporte de Clasificación.
 * Precisión: Para la clase 0 es del 70%, significa que el 70% de las predicciones de la clase 0 fueron correctas, para la clase 1, fueron del 65%.
 * Recall: Para la clase 0 es del 71%, significa que el 71% de las instancias de la clase 0 fueron correctamente identificadas por el modelo. Para la clase 1 fue del 63%
 * F1-sacore: Para la clase 0 es del 71%, que es una medida de la precisión y el recall ponderados. Para la clase 1 es del 64%.
 * Support: Indica el numero de ocurrencias de cada clase en el conjunto de pruebas.

Matriz de confusión.
 * Muestra que el modelo predijo 35 instancias de la clase 0 y 26 de la clase 1. 
 * Predijo incorrectamente 14 instancias de la clase 0 y 15 de la clase 1.

Resumen: El modelo tiene una precisión dl del 67.78% en el conjunto de pruebas.
Aunque el modelo tiene una precisión perfecta en el conjunto de entrenamiento, hay una ligera discrepancia en el rendimiento en los conjuntos de entrenamiento y pruebas, esto nos dice que pudo haber ocurrod u sobreajuste.  

### K-NN (KNeighborsClassifier) (Clasificador KVecinos más cercanos)


```python
X = dataframe.drop('diagnosis', axis=1)
y= dataframe['diagnosis']

X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.3, random_state=42)

model = KNeighborsClassifier()
model.fit(X_train, y_train)
print("Precisión del conjunto de entrenamiento: ", model.score(X_train, y_train))
print("Precisión del conjunto de pruebas: ", model.score(X_test, y_test))

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
matriz_confusion = confusion_matrix(y_test, y_pred)

print('Presición del modelo: {:.2f}%'.format(accuracy*100))

print('Reporte de Clasificación: ')
print(classification_report(y_test, y_pred))

print('Matriz de Confusión: ')
print(confusion_matrix(y_test, y_pred))
sns.heatmap(matriz_confusion, annot=True, fmt='g', cmap='Reds')
```

### Interpretcación de los resultados para K-Vecinos.

Precisión del modelo.
 * La precisión del conjunto de entrenamiento es del 65%
 * La precisión del conjunto de pruebas es del 65.56%.

Reporte de clasificación.
 *  Precisión: Para la clase 0 es del 67%, significa que el 67% de las predicciones de la clase 0 fueron correctas, para la clase 1, fueron del 64%.
 * Recall: Para la clase 0 es del 73%, significa que el 73% de las instancias de la clase 0 fueron correctamente identificadas por el modelo. Para la clase 1 fue del 56%
 * F1-sacore: Para la clase 0 es del 70%, que es una medida de la precisión y el recall ponderados. Para la clase 1 es del 60%.
 * Support: Indica el numero de ocurrencias de cada clase en el conjunto de pruebas.
 
Matriz de confusión.
 * Muestra que el modelo predijo 36 instancias de la clase 0 y 23 de la clase 1. 
 * Predijo incorrectamente 13 instancias de la clase 0 y 18 de la clase 1.
